```python

import numpy as np
import cv2
from classes.RandomPatchMaker import PatchFilter

%load_ext autoreload
%autoreload 2

def tile_statistics_analyzer(patch, const=1):
        #patch = cv2.resize(patch, (56, 56), interpolation=cv2.INTER_LINEAR)
        divider = 8
        filter_size = 56 // divider
        stride = filter_size
        loop_cnt = divider
        cnt_background_tile = 0
        cnt_tissue_like_tile = 0

        max_mean_of_std = 0
        max_R_minus_G = 0

        for i in range(loop_cnt):
            for j in range(loop_cnt):
                if i == 0 or i == loop_cnt - 1 or j == 0 or j == loop_cnt - 1: # 가장자리만 보겠다!
                    x_start = i * stride
                    x_end = i * stride + filter_size
                    y_start = j * stride
                    y_end = j * stride + filter_size
                    check_sample = patch[x_start:x_end,y_start:y_end, :]
         
                    std = np.std(check_sample, axis=(0,1,2))
                    #print(f'{std}:std')
                    mean = np.mean(check_sample, axis=(0,1,2))
                    #print(f'{mean}:mean')
                    
                    if False and mean < 170:
                        plt.imshow(check_sample)
                        plt.title(f'mean:{np.round(mean,2)} / (i:{i},j:{j})')
                        plt.show()
                    
                    R = check_sample[:,:,0]
                    G = check_sample[:,:,1]
                    B = check_sample[:,:,2]

                    R_mean = np.round(np.mean(R),2)
                    G_mean = np.round(np.mean(G),2)
                    B_mean = np.round(np.mean(B),2)
                    
                    mean_value = mean
                    mean = (R_mean, G_mean, B_mean)
                    R_std = np.round(np.std(R),2)
                    G_std = np.round(np.std(G),2)
                    B_std = np.round(np.std(B),2)

                    std = (R_std, G_std, B_std)

                    std_of_mean = np.round(np.std(mean),2) # 작다 => R,G,B 값 모두 비슷 => 배경
                    mean_of_std = np.round(np.mean(std),2) # 작다 => R,G,B의 각 std이 작다 => 배경
                    # print(f'std_of_mean: {std_of_mean}')
                    # print(f'mean_of_std: {mean_of_std}')

                    R_minus_G = np.round(R_mean - G_mean,2)
                    B_minus_G = np.round(B_mean - G_mean,2)
                    # print(f'R_minus_G:{R_minus_G}')
                    # print(f'B_minus_G:{B_minus_G}')

                    if mean_of_std < 2.5 or R_mean - G_mean < 1 or B_mean - G_mean < 1:
                        cnt_background_tile += 1
                    
                    # 180에서 변경
                    if mean_value < 180 and mean_of_std > 30: # 지방도 해당하는 경우 있을 수 있기 때문에 tile의 개수에 대해서도 threshold 적용
                        cnt_tissue_like_tile += 1
                    # 4-A. 
                    if max_mean_of_std < mean_of_std: #그라데이션 마킹도 필터링
                        max_mean_of_std = mean_of_std
                    # 4-B. 
                    if max_R_minus_G < R_minus_G:
                        max_R_minus_G = R_minus_G
                    
        print(f'cnt_background_tile:{cnt_background_tile}/cnt_tissue_like_tile:{cnt_tissue_like_tile}')
        print(f'{max_mean_of_std}')
        print(f'{max_R_minus_G}')
        return False


for patch in patch_list:
    
    patch = cv2.resize(patch, (56, 56), interpolation=cv2.INTER_LINEAR)
    plt.imshow(patch)
    plt.show()
    tile_statistics_analyzer(patch)
    print(PF.is_tissue(patch))
```

```python
import math

from classes.RandomPatchMaker import PatchFilter
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import cv2

%load_ext autoreload
%autoreload 2


PF = PatchFilter()


lists = [not_tissue_patch_list,tissue_patch_list]
del_list= []
if True:
    for j, _list_ in enumerate(lists):
        # Determine the grid size
        num_images = len(_list_)
        grid_size = int(math.ceil(math.sqrt(num_images)))

        # Create a figure to hold the grid
        plt.figure(figsize=(15, 15))  # You can adjust the figure size as needed

        for i, patch in enumerate(_list_):
            ax = plt.subplot(grid_size, grid_size, i + 1)
            
            # Display the image
            plt.imshow(patch)
            
            # Set the title
            is_tissue = PF.is_tissue_test(patch)
            title = 'tissue' if is_tissue else 'not_tissue'
            plt.title(title)

            if j == 0 and title == 'not_tissue':
                del_list.append(i)
            
            # If the patch is an ROI, draw a green rectangle around it
            edgecolor = None
            if is_tissue:
                edgecolor = 'g'
            else:
                edgecolor = 'r'
            # Create a rectangle patch
            rect = patches.Rectangle((0, 0), patch.shape[1], patch.shape[0], linewidth=5, edgecolor=edgecolor, facecolor='none')
            # Add the patch to the Axes
            ax.add_patch(rect)

            
            # Hide axis
            plt.axis('off')

            # Display the complete figure with all subplots
        
        if j == 0:
            plt.suptitle("True Label: not_tissue", color='r',fontsize=40)
        else:
            plt.suptitle("True Label: tissue", color='g',fontsize=40)
        plt.show()

backup_not_tissue_patch_list = not_tissue_patch_list
not_tissue_patch_list = [item for idx, item in enumerate(not_tissue_patch_list) if idx not in set(del_list)]
        
    


```
